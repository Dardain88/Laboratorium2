```python
import random
import time

# Lista produktów (zamiast numerów możesz tu umieścić nazwy produktów)
produkty = [1, 2, 3, 4, 5, 6, 7, 8, 9]

while produkty:
    # Czekamy na wpisanie komendy przez użytkownika
    input("Naciśnij Enter, aby wylosować produkt...")

    # Odliczanie od 5 do 1
    for i in range(5, 0, -1):
        print(f"Losowanie za: {i}")
        time.sleep(1)  # Wstrzymuje wykonanie programu na 1 sekundę

    # Losujemy produkt z dostępnych
    wylosowany_produkt = random.choice(produkty)

    # Wyświetlamy wylosowany produkt
    print(f"\nWylosowano produkt: {wylosowany_produkt}")

    # Usuwamy wylosowany produkt z listy
    produkty.remove(wylosowany_produkt)

    # Wyświetlamy pozostałe produkty do kolejnego losowania
    print(f"Pozostałe produkty: {produkty}\n")

    # Jeśli nie ma już produktów, kończymy pętlę
    if not produkty:
        print("Nie ma więcej produktów do losowania!")
```
